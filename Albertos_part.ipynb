```python
from netCDF4 import Dataset
import numpy as np
import glob
import matplotlib.pyplot as plt

print('list of .nc')
for name in glob.glob('*.nc'):
    print(name)
print('type the file you want to read (without the .nc)')
file=(input())
datas=Dataset(file+'.nc')
var=datas.variables.keys()

Time=datas.variables['Time'][:]

if ['TAT_DI_R' in var]:
    TAT_DI_R=datas.variables['TAT_DI_R'][:]
    TAT_DI_R[TAT_DI_R<0]=['nan']
else:
    TAT_DI_R=[]
    
    
if ['TAT_ND_R' in var]:
    TAT_ND_R=datas.variables['TAT_ND_R'][:]
    TAT_ND_R[TAT_ND_R<0]=['nan']
else:
    TAT_ND_R=[]
    
    
if ['TDEW_GE' in var]:
    TDEW_GE=datas.variables['TDEW_GE'][:]
    TDEW_GE[TDEW_GE<0]=['nan']
else:
    TDEW_GE=[]
    
    
if ['LWC_JW_U' in var]:
    LWC_JW_U=datas.variables['LWC_JW_U'][:]
    LWC_JW_U[LWC_JW_U<0]=['nan']
else:
    LWC_JW_U=[]
    
    
if ['NV_LWC_U' in var]:
    NV_LWC_U=datas.variables['NV_LWC_U'][:]
    NV_LWC_U[NV_LWC_U<0]=['nan']
else:
    NV_LWC_U=[]
    
    
if ['PS_RVSM' in var]:
    PS_RVSM=datas.variables['PS_RVSM'][:]
    PS_RVSM[PS_RVSM<0]=['nan']
else:
    PS_RVSM=[]


    
plt.clf()
plt.close()
plt.figure
plt.figure(figsize=(5,8))
plt.suptitle(file[-4:])
p1=plt.subplot(411)
curve1 = p1.plot(Time,TAT_DI_R.data,label='TAT_DI_R',color='b')
curve2 = p1.plot(Time,TAT_ND_R.data,label='TAT_ND_R',color='r')
curve3 = p1.plot(Time,TDEW_GE.data,label='TDEW_GE',color='g')
#plt.legend(loc='upper left')

p2=plt.subplot(412)
curve4 = p2.plot(Time,LWC_JW_U.data,label='LWC_JW_U',color='b')

p3=plt.subplot(413)
curve5 = p3.plot(Time,NV_LWC_U.data,label='NV_LWC_U',color='b')

p4=plt.subplot(414)
curve5 = p4.plot(Time,PS_RVSM.data,label='PS_RVSM',color='b')

plt.savefig('FAAM.jpg')
plt.show()
```
